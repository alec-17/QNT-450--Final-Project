```python
# creating the function to create a new column with the winner of the game
def match_outcome(row):
    if row['home_team_goal'] > row['away_team_goal']:
        return 'Home Team'
    if row['home_team_goal'] < row['away_team_goal']:
        return 'Away Team'
    if row['home_team_goal'] == row['away_team_goal']:
        return 'Draw'
```

```python
# checking to make sure the function works right 
print(df_Match.apply(lambda row: match_outcome(row), axis=1))
df_Match
```

```python
# applying the function to the 'match' dataframe
df_Match["Match_Winner"] = df_Match.apply(lambda row: match_outcome(row), axis=1)
```

```python
#dropping rows that have 'NA' values
df_Match = df_Match.dropna()
df_Match
```

```python
# creating a function to label BET365's predicted winner
def B365_Pred(row):
    if row['B365H'] < row['B365A']:
        return 'Home Team'
    if row['B365H'] > row['B365A']:
        return 'Away Team'
    if row['B365D'] < row['B365H'] and row['B365D'] < row['B365A']:
        return 'Draw'
    if row['B365H'] == row['B365A']:
        return 'Toss Up'
    
# creating a function to label BW's predicted winner
def BW_Pred(row):
    if row['BWH'] < row['BWA']:
        return 'Home Team'
    if row['BWH'] > row['BWA']:
        return 'Away Team'
    if row['BWD'] < row['BWH'] and row['BWD'] < row['BWA']:
        return 'Draw'
    if row['BWH'] == row['BWA']:
        return 'Toss Up'

# creating a function to label IW's predicted winner
def IW_Pred(row):
    if row['IWH'] < row['IWA']:
        return 'Home Team'
    if row['IWH'] > row['IWA']:
        return 'Away Team'
    if row['IWD'] < row['IWH'] and row['IWD'] < row['IWA']:
        return 'Draw'
    if row['IWH'] == row['IWA']:
        return 'Toss Up'
    
# creating a function to label LB's predicted winner
def LB_Pred(row):
    if row['LBH'] < row['LBA']:
        return 'Home Team'
    if row['LBH'] > row['LBA']:
        return 'Away Team'
    if row['LBD'] < row['LBH'] and row['LBD'] < row['LBA']:
        return 'Draw'
    if row['LBH'] == row['LBA']:
        return 'Toss Up'
    
# creating a function to label WH's predicted winner
def WH_Pred(row):
    if row['WHH'] < row['WHA']:
        return 'Home Team'
    if row['WHH'] > row['WHA']:
        return 'Away Team'
    if row['WHD'] < row['WHH'] and row['WHD'] < row['WHA']:
        return 'Draw'
    if row['WHH'] == row['WHA']:
        return 'Toss Up'
    
# creating a function to label SJ's predicted winner
def SJ_Pred(row):
    if row['SJH'] < row['SJA']:
        return 'Home Team'
    if row['SJH'] > row['SJA']:
        return 'Away Team'
    if row['SJD'] < row['SJH'] and row['SJD'] < row['SJA']:
        return 'Draw'
    if row['SJH'] == row['SJA']:
        return 'Toss Up'
    
# creating a function to label VC's predicted winner
def VC_Pred(row):
    if row['VCH'] < row['VCA']:
        return 'Home Team'
    if row['VCH'] > row['VCA']:
        return 'Away Team'
    if row['VCD'] < row['VCH'] and row['VCD'] < row['VCA']:
        return 'Draw'
    if row['VCH'] == row['VCA']:
        return 'Toss Up'
    
# creating a function to label GB's predicted winner
def GB_Pred(row):
    if row['GBH'] < row['GBA']:
        return 'Home Team'
    if row['GBH'] > row['GBA']:
        return 'Away Team'
    if row['GBD'] < row['GBH'] and row['GBD'] < row['GBA']:
        return 'Draw'
    if row['GBH'] == row['GBA']:
        return 'Toss Up'
    
# creating a function to label BS's predicted winner
def BS_Pred(row):
    if row['BSH'] < row['BSA']:
        return 'Home Team'
    if row['BSH'] > row['BSA']:
        return 'Away Team'
    if row['BSD'] < row['BSH'] and row['BSD'] < row['BSA']:
        return 'Draw'
    if row['BSH'] == row['BSA']:
        return 'Toss Up'
```

```python
# applying the functions listed above
df_Match["B365_Prediction"] = df_Match.apply(lambda row: B365_Pred(row), axis=1)
df_Match["BW_Prediction"] = df_Match.apply(lambda row: BW_Pred(row), axis=1)
df_Match["IW_Prediction"] = df_Match.apply(lambda row: IW_Pred(row), axis=1)
df_Match["LB_Prediction"] = df_Match.apply(lambda row: LB_Pred(row), axis=1)
df_Match["WH_Prediction"] = df_Match.apply(lambda row: WH_Pred(row), axis=1)
df_Match["SJ_Prediction"] = df_Match.apply(lambda row: SJ_Pred(row), axis=1)
df_Match["VC_Prediction"] = df_Match.apply(lambda row: VC_Pred(row), axis=1)
df_Match["GB_Prediction"] = df_Match.apply(lambda row: GB_Pred(row), axis=1)
df_Match["BS_Prediction"] = df_Match.apply(lambda row: BS_Pred(row), axis=1)

df_Match
```

# Calculating Prediction Accuracy of each Betting website

```python
#calcualting B365 prediction accuracy
B365_Home_Team_Correct_Pred = 0
B365_Away_Team_Correct_Pred = 0
B365_Draw_Correct_Pred = 0

for index, row in df_Match.iterrows():
    if row['Match_Winner'] == 'Home Team' and row['Match_Winner'] == row['B365_Prediction']:
        B365_Home_Team_Correct_Pred+=1
    elif row['Match_Winner'] == 'Away Team' and row['Match_Winner'] == row['B365_Prediction']:
        B365_Away_Team_Correct_Pred+=1
    elif row['Match_Winner'] == 'Draw' and row['Match_Winner'] == row['B365_Prediction']:
        B365_Draw_Correct_Pred+=1
    else:
        continue
        
B365_Prediction_Accuracy = ((B365_Home_Team_Correct_Pred + B365_Away_Team_Correct_Pred + B365_Draw_Correct_Pred)
/ len(df_Match.index)) * 100
        
print(f"B365 Correct Prediction when Home Team Wins: {B365_Home_Team_Correct_Pred}")
print(f"B365 Correct Prediction when Away Team Wins: {B365_Away_Team_Correct_Pred}")    
print(f"B365 Correct Prediction when its a Draw: {B365_Draw_Correct_Pred}")
print(f"B365 Prediction Accuracy: {B365_Prediction_Accuracy}%"  )
```
